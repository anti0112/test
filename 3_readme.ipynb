Для этого используется язык разметки Markdown. Он похож на HTML, но еще проще. Взглянем на него.
В Markdown нет специальных скобок и читать его легче. Конечно, сложный текстовый документ с ним не создашь, но нам этого и не нужно.



Небольшой список того, что полезно показать в 'README.md' проекта (md означает Markdown):

* описание,
* возможности,
* пример запуска,
- ссылки на документацию,
- как можно помочь проекту
- и так далее.
