Теперь перейдем к работе с git через консоль, я приведу несколько примеров почему лучше сразу научиться пользоваться git именно через консоль<br>

- Консоль можно использовать с другими редакторами, не только vs code.
- Консольные команды полезны для решения сложных кейсов конфликтов.
- В поисках решения нетипичной задачи на Stack Overflow вы найдете консольную команду.
 

Сейчас мы с помощью командой строки выполним:<br>

- создание репозитория,
- добавление файлов в отлеживаемые (stage),
- создание коммита,
- пуш изменений,
- пул изменений,
- добавление ориджина.

### Инициализация репозитория
Откроем терминал с помощью (CTRL + \`) и напишем команду -> <br>
`git init`<br>
Ответ: Initialized empty Git repository in ...(PATH)

### Добавление файла<br>
Теперь добавим файл 'chapter1.html':

```
Глава 1

И вот однажды с собачкой случилась необычная история. 
День был солнечный и как раз случился выходной.
Она бежала по дорожке и виляла хвостиком, 
так как ей было хорошо.
```

### Проверим статус репозитория<br>
`git status`

### Добавим файлы в отслеживаемые<br>
`git add chapter1.html` — добавит файл<br>
`git add .` — добавит все в текущей папке<br><br>
и снова проверим статус<br>
`git status` 


### Сделаем коммит<br>

Сообщение коммита должно отвечать на вопрос: «Почему были сделаны эти изменения?»<br>

`git commit -m "changed library because outdated"`

Атрибут -m или —message указывает на то, что дальше будет сообщение коммита. Именно оно и следует дальше в кавычках.
Чтобы связать наш локальный репозиторий с репозиторием на GitHub, нам нужно будет авторизовать терминал, так как авторизация, которая была получена для VCS, не будет работать.

### Публикуем проект
С помощью терминала нельзя создать репозиторий на GitHub.
Придется создать его вручную!<br>
#### Создаем репозиторий


#### Теперь привязываем наш репозиторий к репозиторию на GitHub в консоли.

`git remote add origin ВАША ССЫЛКА НА РЕПОЗИТОРИЙ`<br>
Или при создании репозитория у вас появляется окно с командами которые вы можете использовать


### Теперь отправляем все коммиты на сервер (делаем пуш).

`git push origin main`

### Получаем коммиты с сервера
Поменяем что-нибудь на GitHub, например удалим пару строк текста. Теперь версия на сервере опережает локальную. Нужно синхронизироваться!<br>

`git pull`

Когда вы начнете делать коммиты, то столкнетесь с тем, что туда попадают ненужные файлы, например пароли, папки с картинками, файлы настройки вашей IDE, виртуальное окружение (venv) и так далее.

.
У Git есть специальный инструмент, чтобы решить эту проблему, он называется '.gitignore'.

Давайте создадим файлы, которые не будут попадать в коммит. Например, 'mypasswords.py'. Добавляем туда пару строк кода:

```python
superpassword = "qwerty"
```

Добавим файл '.gitignore'. Затем добавим в него название файла. Обратите внимание, у этогого файла нет расширения, хотя это и текстовый файл. Все буквы маленькие и в начале файла стоит точка. Это стандартное имя скрытого системного файла для UNIX-подобных систем.
Откроем '.gitignore' и добавим туда игнорируемый файл.
Мы внутри файла.gitignore
```
mypasswords.py
```

#### Рассмотрим специальный файл для игнорирования указанных файлов git-ом
```
.gitignore
```
